```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df
```

```python
reg = LinearRegression()
X = df.drop('per capita income (US$)', axis = 'columns')
X
y = df['per capita income (US$)']
y
```

```python
reg.fit(X, y)
```

```python
# fig, ax = plt.subplots()
plt.xlabel("Year")
plt.ylabel("Per capita income (US$)")
plt.scatter(df['year'],df['per capita income (US$)'], marker = '*', color = 'green')
plt.plot(df['year'],reg.predict(df[['year']]))
plt.show()
```

```python
fig, ax = plt.subplots()
ax.set_xlabel("Year")
ax.set_ylabel("Per capita income (US$)")
ax.scatter(df['year'],df['per capita income (US$)'], marker = '*', color = 'green')
ax.plot(df['year'],reg.predict(df[['year']]))
plt.show()
```
